```python
import os
import pandas as pd

base_path = "D:/CSV_files" 
all_data = pd.DataFrame()

for month_folder in os.listdir(base_path):
    month_path = os.path.join(base_path, month_folder)
    
    if os.path.isdir(month_path):
        print(f"Processing folder: {month_folder}")
        
       
        for daily_file in os.listdir(month_path):
            daily_file_path = os.path.join(month_path, daily_file)
            
            if daily_file.endswith(".csv"): 
                try:
                    daily_df = pd.read_csv(daily_file_path)
                    date = daily_file.split("_")[0]  
                    daily_df["Date"] = date  
                    all_data = pd.concat([all_data, daily_df], ignore_index=True)
                
                except Exception as e:
                    print(f"Error reading file {daily_file}: {e}")


output_path = "D:/combined_stock_data.csv" 
all_data.to_csv(output_path, index=False)
print(f"All data combined and saved to {output_path}")
```

```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv("D:/Project_Stock/StockData.csv")

# Step 2: Convert stock_date to datetime and extract the month in YYYY-MM format
df['stock_date'] = pd.to_datetime(df['stock_date'])
df['month_year'] = df['stock_date'].dt.to_period('M').astype(str)

# Step 3: Group data by Ticker and month_year, calculating monthly_return as the sum of daily_return
grouped = df.groupby(['Ticker', 'month_year']).agg(monthly_return=('daily_return', 'sum')).reset_index()

# Step 4: For each unique month, get the top 5 gainers and top 5 losers
unique_months = grouped['month_year'].unique()

for month in unique_months:
    # Filter data for the month
    df_month = grouped[grouped['month_year'] == month].copy()
    
    # Identify top 5 gainers (highest monthly_return)
    top_gainers = df_month.nlargest(5, 'monthly_return').copy()
    top_gainers['Category'] = 'Gainer'
    
    # Identify top 5 losers (lowest monthly_return)
    top_losers = df_month.nsmallest(5, 'monthly_return').copy()
    top_losers['Category'] = 'Loser'
    
    # Combine both results
    top_combined = pd.concat([top_gainers, top_losers])
    
    # Step 5: Save the combined data for this month as a CSV file
    output_file = f"Monthly_Returns_{month}.csv"
    top_combined.to_csv(output_file, index=False)
    print(f"Saved {output_file}")
```
