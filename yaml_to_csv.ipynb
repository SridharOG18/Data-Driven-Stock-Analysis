```python
import os
import yaml
import pandas as pd

# Define input and output directories
input_dir = "D:/New Proj/2023-10"  # Folder containing YAML files
output_dir = "D:/CSV_files/2023-10"  # Folder to save CSV files

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Iterate through all YAML files in the input directory
for file_name in os.listdir(input_dir):
    if file_name.endswith(".yaml"):
        input_path = os.path.join(input_dir, file_name)
        
        # Load the YAML file
        with open(input_path, "r") as file:
            data = yaml.safe_load(file)

        # Convert YAML to DataFrame
        df = pd.DataFrame(data)

        # Save the DataFrame as a CSV file
        output_file_name = file_name.replace(".yaml", ".csv")
        output_path = os.path.join(output_dir, output_file_name)
        df.to_csv(output_path, index=False)

        print(f"Converted {file_name} to {output_file_name}")

print("Batch conversion of YAML files to CSV completed!")
```
